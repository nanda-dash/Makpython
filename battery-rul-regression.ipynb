```python
#author : Akshatha Venkatesh Prasad
#date : 30/06/2022
#importing libraries

import numpy as np 
import pandas as pd 
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("/kaggle/input/battery-remaining-useful-life-rul/Battery_RUL.csv")
df['RUL']
```

```python
#finding correlation
corr = df.corr()
sns.heatmap(corr, annot=True)
sns.pairplot(df)
```

```python
#battery ID column creation 
df['Battery ID']= 0 
b=[] 
j=1
for i in df['RUL']: 
    b.append(j) 
    if i == 0: 
        j+=1
        continue
df['Battery ID'] = b 
df.columns.tolist()
```

```python
#splitting dataset into train and test according to battery ID - 9 batteries for training and 5 for testing
train_battery_ids = []
test_battery_ids = []
battery_ids = df['Battery ID'].unique()
for i in battery_ids:
    if i<=8:
        train_battery_ids.append(i)
    else:
        test_battery_ids.append(i)
train_data = df[df['Battery ID'].isin(train_battery_ids)]
test_data = df[df['Battery ID'].isin(test_battery_ids)]
train_data
```

```python
#checking normal distribution of RUL
sns.displot(df['RUL'],kde=True)
```

```python
#dropping cycle ID and Battery ID from the dataset as they don't add value
train_data = train_data.drop('Cycle_Index', axis=1)
test_data = test_data.drop('Cycle_Index', axis=1)
train_data = train_data.drop('Battery ID', axis=1)
test_data = test_data.drop('Battery ID', axis=1)
train_data
```

```python
X_train = train_data.iloc[:,:-1]
y_train = train_data.iloc[:,-1]
X_test = test_data.iloc[:,:-1]
y_test = test_data.iloc[:,-1]
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ax = plt.axes()
ax.scatter(y_test, y_pred, alpha=.5)

ax.set(xlabel='Actual', 
        ylabel='Predictions',
         title='Actual vs Predicted RUL')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

RFR_mseloss = mean_squared_error(y_test, y_pred)
RFR_r2 = r2_score(y_test, y_pred)

RFR_r2
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

ax = plt.axes()
ax.scatter(y_test, y_pred, alpha=.5)

ax.set(xlabel='Actual', 
        ylabel='Predictions',
         title='Actual vs Predicted RUL')
RFR_mseloss = mean_squared_error(y_test, y_pred)
RFR_r2 = r2_score(y_test, y_pred)

RFR_r2
```
