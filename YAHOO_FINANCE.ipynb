```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt

# Set the symbol for the stock
symbol = "AAPL"

# Get the current date
now = datetime.datetime.now()

# Get the date for last year
last_year = now - datetime.timedelta(days=365)

# Get the date for 3 months ago
three_months_ago = now - datetime.timedelta(days=90)

# Get the historical stock data for the last quarter of current year
stock_data_current = yf.download(symbol, start=three_months_ago, end=now, interval='1d')

# Get the historical stock data for the last quarter of previous year
stock_data_last_year = yf.download(symbol, start=last_year - datetime.timedelta(days=90), end=last_year, interval='1d')

# Create the line chart for current year
plt.subplot(2, 1, 1)
plt.plot(stock_data_current['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing price for {} (Current Year)'.format(symbol))

# Create the line chart for last year
plt.subplot(2, 1, 2)
plt.plot(stock_data_last_year['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing price for {} (Last Year)'.format(symbol))

plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
import yfinance as yf
import datetime
from ipywidgets import interact, widgets
import matplotlib.pyplot as plt

def show_stock_data(quarter, year):
    # Get the symbol for the stock
    symbol = "AAPL"
    now = datetime.datetime.now()
    end_date = datetime.datetime(year, (quarter * 3), 1)
    start_date = end_date - datetime.timedelta(days=365*5)
    stock_data = yf.download(symbol, start=start_date, end=end_date, interval='1d')
    plt.plot(stock_data['Close'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title('Closing price for {} for {} of {}'.format(symbol, quarter, year))
    plt.show()

# Create the dropdown menus for selecting the quarter and year
quarter_dropdown = widgets.Dropdown(options=[1, 2, 3, 4], value=1, description='Quarter:',)
year_dropdown = widgets.Dropdown(options=[now.year-i for i in range(5)], value=now.year, description='Year:',)

# Create the button to show the stock data
button = widgets.Button(description="Show")

# Define the function to be called when the button is clicked
button.on_click(lambda b: show_stock_data(quarter_dropdown.value, year_dropdown.value))

# Display the dropdown menus and the button
display(quarter_dropdown)
display(year_dropdown)
display(button)
```

```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
from ipywidgets import widgets, interactive, interact

# Create the quarter dropdown menu
quarters = ['Q1','Q2','Q3','Q4']
quarter_dropdown = widgets.Dropdown(options=quarters, value='Q1', description='Quarter:', layout=widgets.Layout(width='50%'))

# Create the show button
show_button = widgets.Button(description='Show', layout=widgets.Layout(width='10%'))

# Function to handle the show button click
def on_button_click(b):
    symbol = "AAPL"
    now = datetime.datetime.now()
    quarter = quarter_dropdown.value
    months = {'Q1':3,'Q2':6,'Q3':9,'Q4':12}
    month = months[quarter]
    start_date = datetime.datetime(now.year-5, month, 1)
    end_date = datetime.datetime(now.year, month, 1) - datetime.timedelta(days=1)
    stock_data = yf.download(symbol, start=start_date, end=end_date, interval='1d')
    stock_data['Year'] = stock_data.index.year
    stock_data['Quarter'] = stock_data.index.quarter

    # Create the line chart for last 5 years
    plt.figure(figsize=(15,8))
    for year in stock_data.Year.unique():
        data = stock_data[(stock_data.Year == year) & (stock_data.Quarter == months[quarter]//3)]
        plt.subplot(5, 1, year-stock_data.Year.min()+1)
        plt.plot(data['Close'])
        plt.xlabel('Date')
        plt.ylabel('Closing Price')
        plt.title(f'Closing price for {symbol} (Year {year})')
    plt.subplots_adjust(hspace=0.5)
    plt.show()
    
# Attach the function to the button click event
show_button.on_click(on_button_click)

# Display the widgets
widgets.HBox([quarter_dropdown, show_button])
```

```python
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Set the symbol for the stock
symbol = "AAPL"

def stock_data(year,quarter):
    now = datetime.datetime.now()
    start = datetime.datetime(year,quarter*3-2,1)
    end = datetime.datetime(year,quarter*3,1) - datetime.timedelta(days=1)
    end = end if end < now else now
    stock_data = yf.download(symbol, start=start, end=end, interval='1d')
    return stock_data

def plot_data(year,quarter):
    stock_data_current = stock_data(year,quarter)
    plt.figure(figsize=(15,5))
    plt.plot(stock_data_current['Close'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(f'Closing price for {symbol} {year} Q{quarter}')
    plt.show()
    for i in range(1,6):
        stock_data_last = stock_data(year-i,quarter)
        plt.figure(figsize=(15,5))
        plt.plot(stock_data_last['Close'])
        plt.xlabel('Date')
        plt.ylabel('Closing Price')
        plt.title(f'Closing price for {symbol} {year-i} Q{quarter}')
        plt.show()
        

interact(plot_data, year=widgets.IntSlider(min=2000,max=2022,value=2020),
         quarter=widgets.IntSlider(min=1,max=4,value=1))
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Create a text box for the user to enter the stock symbol
symbol_textbox = widgets.Text(value='AAPL', description='Stock Symbol:')

def stock_data(year,quarter, symbol):
    now = datetime.datetime.now()
    start = datetime.datetime(year,quarter*3-2,1)
    end = datetime.datetime(year,quarter*3,1) - datetime.timedelta(days=1)
    end = end if end < now else now
    stock_data = yf.download(symbol, start=start, end=end, interval='1d')
    return stock_data

def plot_data(year,quarter, symbol):
    stock_data_current = stock_data(year,quarter, symbol)
    p = figure(title=f'Closing price for {symbol} {year} Q{quarter}', x_axis_type='datetime', width=1000, height=300)
    p.line(x=stock_data_current.index, y=stock_data_current["Close"])
    show(p)
    for i in range(1,6):
        stock_data_last = stock_data(year-i,quarter, symbol)
        p = figure(title=f'Closing price for {symbol} {year-i} Q{quarter}', x_axis_type='datetime', width=1000, height=300)
        p.line(x=stock_data_last.index, y=stock_data_last["Close"])
        show(p)
        
output_notebook()
# Pass the stock symbol textbox to the interact function as an argument
interact(plot_data, year=widgets.IntSlider(min=2000,max=2022,value=2020),
         quarter=widgets.IntSlider(min=1,max=4,value=1), symbol=symbol_textbox)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Create a text box for the user to enter the stock symbol
symbol_textbox = widgets.Text(value='AAPL', description='Stock Symbol:')

def stock_data(year,quarter, symbol):
    now = datetime.datetime.now()
    start = datetime.datetime(year,quarter*3-2,1)
    end = datetime.datetime(year,quarter*3,1) - datetime.timedelta(days=1)
    end = end if end < now else now
    stock_data = yf.download(symbol, start=start, end=end, interval='1d')
    return stock_data

def plot_data(year,quarter, symbol):
    stock_data_current = stock_data(year,quarter, symbol)
    stock_data_current["Change"] = stock_data_current["Close"].pct_change()
    p = figure(title=f'Closing price for {symbol} {year} Q{quarter}', x_axis_type='datetime', y_axis_label='Percentage Change', width=1000, height=300)
    p.line(x=stock_data_current.index, y=stock_data_current["Change"]*100)
    show(p)
    for i in range(1,6):
        stock_data_last = stock_data(year-i,quarter, symbol)
        stock_data_last["Change"] = stock_data_last["Close"].pct_change()
        p = figure(title=f'Closing price for {symbol} {year-i} Q{quarter}', x_axis_type='datetime', y_axis_label='Percentage Change', width=1000, height=300)
        p.line(x=stock_data_last.index, y=stock_data_last["Change"]*100)
        show(p)
        
output_notebook()
# Pass the stock symbol textbox to the interact function as an argument
interact(plot_data, year=widgets.IntSlider(min=2000,max=2022,value=2020),
         quarter=widgets.IntSlider(min=1,max=4,value=1), symbol=symbol_textbox)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Create a text box for the user to enter the stock symbol
symbol_textbox = widgets.Text(value='AAPL', description='Stock Symbol:')

def stock_data(year,quarter, symbol):
    now = datetime.datetime.now()
    start = datetime.datetime(year,quarter*3-2,1)
    end = datetime.datetime(year,quarter*3,1) - datetime.timedelta(days=1)
    end = end if end < now else now
    stock_data = yf.download(symbol, start=start, end=end, interval='1d')
    return stock_data

def plot_data(year,quarter, symbol):
    stock_data_current = stock_data(year,quarter, symbol)
    stock_data_current["Change"] = (stock_data_current["Close"] / stock_data_current["Close"].shift(1)) - 1
    p = figure(title=f'Closing price for {symbol} {year} Q{quarter}', x_axis_type='datetime', y_axis_label='% Gain or Loss', width=1000, height=300)
    p.line(x=stock_data_current.index, y=stock_data_current["Change"]*100)
    show(p)
    for i in range(1,6):
        stock_data_last = stock_data(year-i,quarter, symbol)
        stock_data_last["Change"] = (stock_data_last["Close"] / stock_data_last["Close"].shift(1)) - 1
        p = figure(title=f'Closing price for {symbol} {year-i} Q{quarter}', x_axis_type='datetime', y_axis_label='% Gain or Loss', width=1000, height=300)
        p.line(x=stock_data_last.index, y=stock_data_last["Change"]*100)
        show(p)

output_notebook()
# Pass the stock symbol textbox to the interact function as an argument
interact(plot_data, year=widgets.IntSlider(min=2000,max=2022,value=2020),
         quarter=widgets.IntSlider(min=1,max=4,value=1), symbol=symbol_textbox)
```
